```
num_missing = nulldata.sum(axis=1)
index_for_null_rows = list(num_missing[num_missing == 21].index)
print("Number of null rows: ", len(index_for_null_rows))
print("Percentage of null rows: ", (len(index_for_null_rows)/len(data.axes[0])*100))
print("index for null rows: ", *index_for_null_rows)
```
